```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L   
```

movies['genres'].apply(convert)

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres']
```

```python
def convert(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L   
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies['cast']
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] =='Director':
            L.append(i['name'])
            break
    return L   
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['crew']
```

```python
 movies['overview']= movies['overview'].str.split()
```

```python
 movies['overview']
```

```python
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['crew']+movies['cast']+movies['genres']
```

```python
movies.head()
```

```python
df = movies[['id','title','tags']]
df.head()
```

```python
def convert_to_list(x):
    if isinstance(x, float) and pd.isna(x):
        return []
    elif isinstance(x, list):
        return x
    else:
        return [x]

# Step 2: Apply the function to the 'tags' column
df['tags'] = df['tags'].apply(convert_to_list)

# Step 3: Ensure all elements in the lists are strings
df['tags'] = df['tags'].apply(lambda x: [str(i) for i in x])

# Step 4: Join the elements in the lists with a space
df['tags'] = df['tags'].apply(lambda x: " ".join(x))

print(df)
```

```python
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key=lambda x:x[1])
```

```python
df[df['title'] == 'Avatar'].index[0]
```

```python
def recommend(movie):
    movies_index = df[df['title'] == movie].index[0]
    distance = similarity[movies_index]
    movies_list = sorted(list(enumerate(distance)),reverse = True,key=lambda x:x[1])[1:8]
    
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
